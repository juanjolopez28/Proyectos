```python
import pandas as pd
#Carga de los datos sin provincia o localidad
#Eliminación de columnas que se duplican


#Accesos a internet por cada 100 hogares por trimestre [Año,Trimestre,Accesos por cada 100 hogares,Accesos por cada 100 hab,Periodo]
df_internet_100h=pd.read_csv('Datasets/1_internet_Penetracion_100hogares_sinprovincia.csv',delimiter=',',header='infer')

#Acesos a Internet banda ancha y angos por trimestre [Banda ancha fija,Dial up,Total]
df_internet_banda_ancha_rangos=pd.read_csv('Datasets/3_internet_bandaAncha_angosta_sinprovincia.csv',delimiter=',',header='infer')
df_internet_banda_ancha_rangos.head
df_internet_banda_ancha_rangos.drop(['Año', 'Trimestre','Periodo'], axis=1, inplace=True)

#Acceso a internet por tecnología por trimestre [ADSL,Cablemodem,Fibra optica,Wireless,Otros,Total]
df_internet_tecnologia=pd.read_csv('Datasets/5_internet_accesos-por-tecnologia_sinprovincia.csv',delimiter=',',header='infer')
df_internet_tecnologia.drop(['Año', 'Trimestre','Periodo'], axis=1, inplace=True)


#Velocidad de bajada media o promedio por trimestre[Mbps (Media de bajada)]
df_internet_velocidad_media=pd.read_csv('Datasets/8_velocidad_debajada_media_internet_sinprovincia.csv',delimiter=',',header='infer')
#no tiene periodo
df_internet_velocidad_media.drop(['Año', 'Trimestre'], axis=1, inplace=True)


#Velocidad de bajada ´por rango [Entre 512 Kbps y 1 Mbps,Entre 1 Mbps y 6 Mbps , etc]
df_internet_velocidad_rango=pd.read_csv('Datasets/10_internet_accesos-por-velocidad_rango_sinprovincia.csv',delimiter=',',header='infer')
#no tiene periodo
df_internet_velocidad_rango.drop(['Año', 'Trimestre'], axis=1, inplace=True)
df_internet_velocidad_rango = df_internet_velocidad_rango.drop(df_internet_velocidad_rango.columns[-1], axis=1)


#Ingresos de internet por trimestre [Ingresos (miles de pesos)]
df_internet_ingresos=pd.read_csv('Datasets/14_internet_Ingresos.csv',delimiter=',',header='infer')
df_internet_ingresos.drop(['Año', 'Trimestre','Periodo'], axis=1, inplace=True)
df_internet_ingresos.head


```

```python
#Cambio de tipo de datos y correción de formato
#Accesos a internet por cada 100 hogares por trimestre [Año,Trimestre,Accesos por cada 100 hogares,Accesos por cada 100 hab,Periodo]
df_internet_100h['Accesos por cada 100 hogares']=df_internet_100h['Accesos por cada 100 hogares'].str.replace(',', '.',regex=False).astype(float)

df_internet_100h['Accesos por cada 100 hab']=df_internet_100h['Accesos por cada 100 hab'].str.replace(',', '.',regex=False).astype(float)
df_internet_100h.dtypes
```

```python
#Acesos a Internet banda ancha y angos por trimestre [Banda ancha fija,Dial up,Total]
df_internet_banda_ancha_rangos['Banda ancha fija']=df_internet_banda_ancha_rangos['Banda ancha fija'].str.replace('.', '',regex=False).astype(int)
#en el caso de accaesos por dial up lo carga como int por eso debemos cambiar
df_internet_banda_ancha_rangos['Dial up']=df_internet_banda_ancha_rangos['Dial up'].astype(str).str.replace('.', '',regex=False).astype(int)
df_internet_banda_ancha_rangos['Total']=df_internet_banda_ancha_rangos['Total'].str.replace('.', '',regex=False).astype(int)
df_internet_banda_ancha_rangos.rename(columns={'Total': 'Total_BA'}, inplace=True)
df_internet_banda_ancha_rangos.dtypes
df_internet_banda_ancha_rangos.head
```

```python
#Acceso a internet por tecnología por trimestre [ADSL,Cablemodem,Fibra optica,Wireless,Otros,Total]
df_internet_tecnologia.dtypes
df_internet_tecnologia['ADSL']=df_internet_tecnologia['ADSL'].str.replace('.','',regex=False).astype(int)
df_internet_tecnologia['Cablemodem']=df_internet_tecnologia['Cablemodem'].str.replace('.','',regex=False).astype(int)
df_internet_tecnologia['Fibra óptica']=df_internet_tecnologia['Fibra óptica'].str.replace('.','',regex=False).astype(int)
df_internet_tecnologia['Wireless']=df_internet_tecnologia['Wireless'].astype(str).str.replace('.','',regex=False).astype(int)
df_internet_tecnologia['Otros']=df_internet_tecnologia['Otros'].astype(str).str.replace('.','',regex=False).astype(int)
df_internet_tecnologia['Total']=df_internet_tecnologia['Total'].str.replace('.','',regex=False).astype(int)
df_internet_tecnologia.rename(columns={'Total': 'Total_tecno'}, inplace=True)
df_internet_tecnologia.dtypes
```

```python
#Velocidad de bajada media o promedio por trimestre[Mbps (Media de bajada)]
df_internet_velocidad_media.dtypes
df_internet_velocidad_media['Mbps (Media de bajada)']=df_internet_velocidad_media['Mbps (Media de bajada)'].str.replace(',','.').astype(float)
df_internet_velocidad_media.dtypes
```

```python
#Ingresos de internet por trimestre [Ingresos (miles de pesos)]
df_internet_ingresos['Ingresos (miles de pesos)']=df_internet_ingresos['Ingresos (miles de pesos)'].str.replace('.','',regex=False).astype(int)
```

```python
#Realizamos nuestra tablas para trabajar con la data

#1.-Tabla de tiempo tendra los siguientes atributos: ID_tiempo, Año, Trimestre, Periodo

df_tiempo=df_internet_100h[['Año','Trimestre','Periodo']]
df_tiempo.insert(0, 'ID_tiempo', range(1, len(df_internet_100h) + 1))
df_internet_100h.drop(['Año', 'Trimestre','Periodo'], axis=1, inplace=True)
```

```python
#2.- Tabla para datos trimestrales
df_datos_trimestrales=pd.concat([df_tiempo['ID_tiempo'],df_internet_100h,df_internet_banda_ancha_rangos,df_internet_tecnologia,df_internet_velocidad_media,df_internet_velocidad_rango,df_internet_ingresos],axis=1)
df_datos_trimestrales.insert(0, 'ID_datos_tri', range(1, len(df_internet_100h) + 1))
df_datos_trimestrales
```
